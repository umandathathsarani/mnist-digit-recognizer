# Exploring the Hidden MNIST Dataset

Since my AI struggled a bit with my custom drawing, I want to physically see the images it learned from. 

TensorFlow downloaded the dataset automatically during training, but it saved them as raw data arrays instead of normal image files. In this notebook, I am going to convert those arrays back into standard `.png` images so I can study the line thickness, the drawing styles, and the black padding around the numbers.

```python
import tensorflow as tf
from PIL import Image
import os
```

First, I need to load the dataset into memory exactly like I did in my training script. 

I only need the training images (`x_train`) and their correct labels (`y_train`) for this experiment. I can safely ignore the testing data by assigning it to an underscore `_`.

```python
(x_train, y_train), _ = tf.keras.datasets.mnist.load_data()
```

Now, I am going to write a script to create a new folder called `sample_images` right here in my project directory. 

Then, I will loop through the very first 10 items in the dataset. For each item, I will grab the grid of numbers, turn it back into visual pixels using the Python Imaging Library (`PIL`), and save it to the folder. I am also adding the actual number to the file name so I know exactly what the AI was told it represents!

```python
# Create the folder if it doesn't exist yet
os.makedirs("sample_images", exist_ok=True)

# Loop through the first 10 images in the dataset
for i in range(10):
    image_array = x_train[i]      # The raw pixel data
    number_label = y_train[i]     # The correct answer 
    
    # Convert the array into a standard image object
    img = Image.fromarray(image_array)
    
    # Create a clear file name and save it
    filename = f"sample_images/digit_{number_label}_index_{i}.png"
    img.save(filename)

print("Images saved successfully! Time to check the 'sample_images' folder on the left.")
```
