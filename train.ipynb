# MNIST Handwritten Digit Recognizer
This is my first deep learning project. The goal here is to train an AI to recognize human handwriting, specifically numbers from 0 to 9. 

First, we need to import our main tool, TensorFlow, and Matplotlib so we can draw graphs if we want to.

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
```

Next, we need to load our data. The MNIST dataset is already built into TensorFlow, which makes things easy. 

Computers like small numbers, so we are going to divide all the pixel values by 255.0. This squishes the image data down to a scale between 0 and 1, which helps the AI learn much faster.

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train, x_test = x_train / 255.0, x_test / 255.0
```

Now we build the actual "brain" of our AI. 
- The `Flatten` layer takes our 28x28 pixel square image and stretches it out into a single long line of pixels.
- The `Dense` layer with 128 nodes is the hidden part of the brain that looks for patterns.
- The `Dropout` layer randomly turns off 20% of the brain during training so it doesn't just memorize the answers.
- The final `Dense` layer has 10 nodes, one for each number (0 through 9). The 'softmax' part turns the output into percentages, so the AI can say "I am 99% sure this is a 7".

```python
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])
```

Before we can train the model, we have to compile it. This means giving the AI a way to measure how wrong it is (the 'loss' function) and a way to adjust itself to get better (the 'adam' optimizer). We also tell it that we care about 'accuracy'.

Then, we start the training! We will run it through our dataset 5 times (epochs).

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=5)
```

The training is done. Now, we need to test it on the `x_test` images it has never seen before to make sure it actually learned how to read digits, rather than just memorizing the training test. 

Finally, we save the trained brain into a file so we don't have to train it all over again next time we want to use it.

```python
model.evaluate(x_test,  y_test, verbose=2)

model.save('mnist_model.keras')
```
