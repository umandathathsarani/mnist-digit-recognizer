# Extracting the Complete MNIST Dataset

I want to extract all 70,000 images from the hidden NumPy arrays and save them as actual `.png` files on my computer. 

Because 70,000 files is a massive amount of data, I am going to write a script that creates a master folder called `mnist_full_dataset`, and inside that, creates 10 subfolders (`0` through `9`). It will sort every single image into its correct folder based on its label.

```python
import tensorflow as tf
from PIL import Image
import os
import numpy as np
```

First, I need to load both the training data (60,000 images) and the testing data (10,000 images). Then, I will use `numpy` to stitch them together into one massive list of 70,000 images so I can process them all in one go.

```python
# Load all the data
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Combine training and testing sets
all_images = np.concatenate((x_train, x_test))
all_labels = np.concatenate((y_train, y_test))

print(f"Total images to process: {len(all_images)}")
```

Next, I will set up the folder structure. This code will create the main directory and the 10 subdirectories for the numbers.

```python
base_dir = "mnist_full_dataset"
os.makedirs(base_dir, exist_ok=True)

# Create subfolders for digits 0-9
for i in range(10):
    os.makedirs(os.path.join(base_dir, str(i)), exist_ok=True)
    
print("Folders created successfully.")
```

Finally, the big loop. This will take a few minutes to run. I am adding a simple print statement that updates me every time it finishes 10,000 images so I know it hasn't crashed.

```python
print("Starting extraction... this might take a minute or two!")

for i in range(len(all_images)):
    # Grab the data
    image_array = all_images[i]
    label = all_labels[i]
    
    # Convert to an image
    img = Image.fromarray(image_array)
    
    # Construct the file path (e.g., mnist_full_dataset/5/img_402.png)
    filepath = os.path.join(base_dir, str(label), f"img_{i}.png")
    
    # Save it
    img.save(filepath)
    
    # Print a progress update
    if (i + 1) % 10000 == 0:
        print(f"Processed {i + 1} images out of 70000...")

print("Extraction complete! Check your file explorer.")
```
